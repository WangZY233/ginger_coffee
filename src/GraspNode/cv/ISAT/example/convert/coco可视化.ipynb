# coco中存在两种存储方式，顶点存储方式与rle存储方式
- 顶点存储方式，存储了多边形的顶点，具体格式为:[x1,  y1, x2, y2, ..., xn, yn]
- rle格式，主要用于存储标记为iscrowd的目标，需要通过pycocotools库转换为mask进行处理

```python
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
from PIL import Image
import random
import os
```

```python
COCOROOT = 'XXX'
SET = 'val2017'
```

```python
coco = COCO(os.path.join(COCOROOT, 'annotations/instances_{}.json'.format(SET)))
# has_crowd_ids = [img_id for img_id in coco.imgs if sum(anno['iscrowd'] for anno in coco.loadAnns(coco.getAnnIds(img_id)))]
# random_ids = random.choices(list(coco.imgs.keys()), k=3)
# random_ids = random_ids + random.choices(has_crowd_ids, k=2)

random_ids = [83540, 76261, 13923, 549930, 100723]
```

```python
random_ids
```

```python
for id in random_ids:
    imgs = coco.loadImgs(id)
    anno_id = coco.getAnnIds(id)
    anno = coco.loadAnns(anno_id)
    img_path = os.path.join(os.path.join(COCOROOT, SET), imgs[0]['file_name'])
    img = Image.open(img_path)
    plt.imshow(img)
    coco.showAnns(anno)
    plt.show()
```
